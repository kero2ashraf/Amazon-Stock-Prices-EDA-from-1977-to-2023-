# Some EDA Visualizations

```python
data["Market Cap"] = data["Open"]*data["Volume"]
```

```python
#Date vs Market cap
fig,ax=plt.subplots(figsize=(20,8))
ax.plot(data['Date'],data['Market Cap'],color="Black")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date',fontsize='14')
ax.set_ylabel('Market Cap',fontsize='14')
plt.title('Market Cap',fontsize='14')
plt.grid()
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(20,8))
ax.bar(data['Date'],data['Market Cap'],color="Orange")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date',fontsize='14')
ax.set_ylabel('Market Cap',fontsize='14')
plt.title('Market Cap',fontsize='14')
plt.grid()
plt.show()
```

```python
data.iloc[data["Market Cap"].argmax()]
```

```python
ohlc = data[(data["Date"]>"1997-05-15")]
ohlc=ohlc.loc[:,["Date","Open","High","Low","Close","Volume","Market Cap"]]
ohlc["Date"]=pd.to_datetime(ohlc["Date"],format="%Y-%m-%d")
fig,ax=plt.subplots(figsize=(20,8))
ax.plot(ohlc["Date"],ohlc["Market Cap"],color="Orange")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel("Date",fontsize="14")
ax.set_ylabel("Market Cap",fontsize="14")
plt.title("Market Cap",fontsize="14")
plt.grid()
plt.title("Market Cap(After 15 may , 1977)",fontsize="20")
plt.show()
```

```python
data["Vol"]=(data["Close"]/data["Close"].shift(1))-1
```

```python
#Date vs Vol 
fig,ax=plt.subplots(figsize=(20,8))
ax.plot(data["Date"],data["Vol"],color="purple")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title("Voltatility",fontsize="14")
plt.grid()
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(20,8))
ax.bar(data["Date"],data["Vol"],color="purple")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title("Voltatility",fontsize="14")
plt.grid()
plt.show()
```

```python
data["Vol"].hist(bins=1000,color="Blue");
```

```python
data["Cumulative Return"] = (1+data["Vol"]).cumprod()
```

```python
#Date vs Cumulative Return
fig,ax=plt.subplots(figsize=(20,8))
ax.bar(data["Date"],data["Cumulative Return"],color="Brown")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel("Date",fontsize="14")
ax.set_ylabel("Cumulative Return",fontsize="14")
plt.title("Cumulative Return",fontsize="14")
plt.grid()
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(20,8))
ax.plot(data["Date"],data["Cumulative Return"],color="Brown")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel("Date",fontsize="14")
ax.set_ylabel("Cumulative Return",fontsize="14")
plt.title("Cumulative Return",fontsize="14")
plt.grid()
plt.show()
```

```python
ohlc = data[(data['Date'] > '1997-05-15')]
ohlc = ohlc.loc[:, ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap',"Vol",
       'Cumulative Return']]
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(ohlc['Date'], ohlc['Cumulative Return'], color='green')
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Cumulative Return', fontsize='11')
plt.grid()
plt.title('Cumulative Return (After 05 May, 1997)', fontsize='20')
plt.show()
```
